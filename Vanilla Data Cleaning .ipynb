```python
lens=df['New Flowers'].str.split(',').map(len)
```

## Values succesfully edited. And no mismatch!

```python
newdf=pd.DataFrame({'Date':np.repeat(df['Date'],lens),
                   'Location':np.repeat(df['Location'],lens),
                    'Points':np.repeat(df['Points'],lens),
                    'New Flowers': chainer(df['New Flowers']),
                    'Flower total': chainer(df['Flower total'])
                    })
```
